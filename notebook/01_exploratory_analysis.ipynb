Price.DE. : The price of the electric vehicle in Germany.

As we can see the Price.DE. is the price of EVs in Germany , So we need to convert it into USD because our project is based on US datasets

### 2.1.3 Converting the German Price into USD

```python
exchange_rate = 1.08 
df1['price_usd_estimated'] = df1['price.de.'] * exchange_rate
df1['price_usd_estimated'] = df1['price_usd_estimated'].round(0).astype('Int64')
```

```python
df1.head(3)
```

Now , the new column is added which is the EVs price in USD after conversion

```python
df1.isnull().sum()
```

```python
df1['fast_charge']=df1['fast_charge'].fillna(df1['fast_charge'].median())
df1['price.de.']=df1['price.de.'].fillna(df1['price.de.'].median())
df1['price_usd_estimated']=df1['price_usd_estimated'].fillna(df1['price_usd_estimated'].median())
```

### 2.1.4 Dropping the unnecessary columns

```python
df1.drop('car_name', axis=1, inplace=True)
df1.drop('car_name_link', axis=1, inplace=True)
```

### 3.3.2 One Hot Encoding

```python
df3=pd.get_dummies(df3,columns=['vehicle_model'] , drop_first=False, dtype='int')
```

```python
df3=pd.get_dummies(df3,columns=['charging_station_location'] , drop_first=False, dtype='int')
```

```python
df3 = pd.get_dummies(df3, columns=['user_type'],drop_first=False, dtype='int')
```

### 3.3.3 Applying Ordinal Encoding on object type columns
 (bcoz they r having ordinal categorical data values)

```python
order =['Morning', 'Afternoon', 'Evening','Night']
df3['time_of_day'] = pd.Categorical(df3['time_of_day'], categories=order, ordered=True)
df3['time_of_day'] = df3['time_of_day'].cat.codes
```

```python
order =['Sunday','Monday', 'Tuesday', 'Wednesday','Thursday', 'Friday', 'Saturday']
df3['day_of_week'] = pd.Categorical(df3['day_of_week'], categories=order, ordered=True)
df3['day_of_week'] = df3['day_of_week'].cat.codes
```

```python
order =['Level 1', 'Level 2', 'DC Fast Charger']
df3['charger_type'] = pd.Categorical(df3['charger_type'], categories=order, ordered=True)
df3['charger_type'] = df3['charger_type'].cat.codes
```

### 3.3.4 Converting required columns into datetime format

```python
df3['charging_start_time'] = pd.to_datetime(df3['charging_start_time'], unit='s')
df3['charging_end_time'] = pd.to_datetime(df3['charging_end_time'], unit='s')
```

### 3.3.5 Taking out the specific details and creating separate columns

```python
df3['start_hour'] = df3['charging_start_time'].dt.hour
df3['start_day'] = df3['charging_start_time'].dt.day
df3['start_month'] = df3['charging_start_time'].dt.month

df3['end_hour'] = df3['charging_end_time'].dt.hour
df3['end_day'] = df3['charging_end_time'].dt.day
df3['end_month'] = df3['charging_end_time'].dt.month
```

```python
df3.drop(['charging_start_time','charging_end_time'], axis=1, inplace=True)
```
